```python
import pandas as pd
from datetime import datetime
```

```python
df = pd.read_csv(r'D:\AI Engineer\Udemy - Data Science _ Machine Learning(Theory+Projects)A-Z 90 HOURS\2. Python for Data Understanding and Data Visualization\covid_19_till_19_May_2020.csv')
```

```python
df.head()
```

```python
m = folium.Map()
df['Date'] = pd.to_datetime(df['Date'])
df2 = df[df['Date']==max(df['Date'])]
```

```python
df2
```

```python
for i in range(0,len(df2)):
    folium.Circle(location=[df2.iloc[i]['Lat'],df2.iloc[i]['Long']],
                  color='crimson',fill='crimson',
                  radius=int(df2.iloc[i]['Confirmed'])).add_to(m)
m
```
