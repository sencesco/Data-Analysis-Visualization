### Axis limits

```python
fig = plt.figure(figsize=(10,2.5))

plt.subplot(1,2,1)
plt.plot(x,y1,'-.k')
plt.plot(x,y2,':g')
plt.plot(x,y3,'--r')
plt.xlim(-5,15)
plt.ylim(-3,2.5)

plt.subplot(1,2,2)
plt.plot(x,y1,'-.k')
plt.plot(x,y2,':g')
plt.plot(x,y3,'--r')
# this way also same as take seperate of pli.xlim() and plt.ylim()
plt.axis([-5,15,-3,2.5])
```

```python
# We can set the resolution of a graph with dpi
fig = plt.figure(figsize=(7.3,2), dpi=120)

plt.subplot(1,2,1)
plt.plot(x,y1,'-.k')
plt.axis('equal')

plt.subplot(1,2,2)
plt.plot(x,y1,'-.k')
plt.axis('tight')
```

```python
# we can use plt.xlim and plt.ylim() to reverse a graph
fig = plt.figure(figsize=(5,1.5), dpi=120)

plt.subplot(1,2,1)
plt.plot(x,2*x+3,'-.k')
plt.xlim(0,10)

plt.subplot(1,2,2)
plt.plot(x,2*x+3,'-.k')
plt.xlim(10,0)
```

```python
fig = plt.figure(figsize=(3.3,2))
plt.plot(x,y1,'-.r')
plt.xlim(10,0)
plt.ylim(1.5,-1.5)
```

### Contour plots

- X, Y: These parameter are the coordinates of the values in Z.
- Z : This parameter is the height values over which the contour is drawn.
- levels : This parameter is used to determine the numbers and positions of the contour lines / regions.
- z = f(x,y)  that any f(x,y) point is the same z that mean there are same height or same level.

```python
# Return coordinate matrices from coordinate vectors.

X,Y = np.meshgrid([-1,1],[5,8])
```

```python
X
```

```python
Y
```

$\;\;\;\;$ The result will -1 join 5, -1 join 8, 1 join 5 and 1 joint 8 that is a meshgrid boundary

<div></div>

```python
x = np.linspace(0,5,50)
y = np.linspace(0,5,40)

X,Y = np.meshgrid(x,y)
```

```python
X
```

```python
Z = 10*np.sin(X)+2*Y
```

$\;\;\;\;$ Contour with not fill

```python
fig = plt.figure(figsize=(12,3))

# Contour with not fill plt.contour()
plt.subplot(1,2,1)
plt.contour(X,Y,Z,30,cmap='RdGy')
plt.colorbar()
plt.xlabel('X')
plt.ylabel('Y')
plt.title('contour plot')

# Contour with fill plt.contourf)
plt.subplot(1,2,2)
plt.contourf(X,Y,Z,30,cmap='RdGy')
plt.colorbar()
plt.xlabel('X')
plt.ylabel('Y')
plt.title('contour plot')
```

### Histograms

```python
d1 = np.random.normal(0,0.7,1000)
d2 = np.random.normal(-5,2,1000)
d3 = np.random.normal(5,3,1000)

plt.figure(figsize=(5,3))
plt.hist(d1, bins = 50)
plt.hist(d2, bins = 50, alpha = 0.5)
plt.hist(d3, bins = 50, alpha = 0.75)
plt.show()
```

```python
plt.figure(figsize=(5,3))
params = dict(histtype = 'stepfilled', alpha =0.4, bins = 30)

plt.hist(iris.sepal_length, **params)
plt.hist(iris.sepal_width, **params)
plt.hist(iris.petal_length, **params)
plt.hist(iris.petal_width, **params)
plt.show()
```

### 3D Plot

```python
# this toll for 3D plot

from mpl_toolkits.mplot3d import Axes3D
```

```python
plt.style.use('fivethirtyeight')

# ax = fig.gca(projection='3d') This is old version of matplotlib and can not use anymore
ax =plt.figure(figsize=(7,4)).add_subplot(projection='3d')

r = np.linspace(-2,2,100)**2+1
theta = np.linspace(-4*np.pi,4*np.pi,100)
X = r*np.sin(theta)
Y = r*np.cos(theta)
Z = np.linspace(-2,2,100)
ax.plot(X,Y,Z,label="3D curve")

ax.legend()        
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Plot')
```
