# Seaborn

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
plt.style.use('ggplot')
```

```python
df = pd.DataFrame({'X': np.random.normal(0,2,1000),
                   'Y': np.random.normal(3,5,1000),
                   'z': np.random.normal(-4,10,1000)})
```

```python
df.head()
```

### Histograms

```python
# Defualt style from seaborn
sns.set()
label=['X','Y','Z']
i = 0
plt.figure(figsize=(7,4))
for col in df.columns:
    plt.hist(df[col],density=True,alpha=0.5, bins=30, label=label[i])
    sns.kdeplot(df[col],fill=True)
    plt.legend()
    i+=1
```

### Joint 2 variable plot 

```python
plt.figure(figsize=(7,4))
with sns.axes_style('white'):
    sns.jointplot(data=df,x='X',y='Y',kind='kde', fill=True)
```

### Heatmaps

```python
df = pd.DataFrame(np.random.rand(10,10), columns=list('ABCDEFGHIJ'))
```

```python
df
```

```python
sns.heatmap(df,annot=True,annot_kws={'size':7})
```

$\;\;\;\;$ And most of the cases when we generate confusion, matrices in machine learning, for evolution, If if the number of classes in the classification dataset is large, the confusion matrix is normally a big matrix. And one way to just visualize the results is just to create a heat map that give a very quick description of what is happening behind.
