```python
y
```

```python
# this is same thing
y = np.add(A,10)
y = A+10
```

```python
z = np.zeros(10)

np.add(A,10,out=z[::2])

# out=z[::2]): Adds 10 to each element of the array A. 
# The out=z[::2] part specifies that the result should be stored in the array z (locate im this memory)
# starting from the first element (index 0) and going by steps of 2 (every other element).

# This is benefit when sometimes working with the large array and inted to avoid the temporary storage.
```

```python
z
```

```python
z[::2] = A+50
```

```python
z
```

```python
fig, ax = plt.subplots(1,3)

ch = ['Red','Green','Blue']

# axes.flat is not a function, it's an attribute of the numpy.ndarray: that can iteration 
# the np.array from 1- n dimension of the array.
for i,axi in enumerate(ax.flat):
    
    axi.imshow(img[:,:,i],plt.cm.gray)
    axi.set(xticks=[],yticks=[])
    axi.set_xlabel(ch[i],color=ch[i].lower())
    fig.set_size_inches(20,50)
```

$\;\;\;\;$ ref of axes.flat: https://stackoverflow.com/questions/46862861/what-does-axes-flat-in-matplotlib-do

```python
# For example increasing the red content

fig, ax = plt.subplots(1,3)

img2 = img.copy()
img2[:,:,2] = img[:,:,0]*10

ch = ['Red','Green','Blue']
for i,axi in enumerate(ax.flat):
    
    axi.imshow(img2[:,:,i],plt.cm.gray) # for increasing the red content
    axi.set(xticks=[],yticks=[])
    axi.set_xlabel(ch[i],color=ch[i].lower())
    fig.set_size_inches(20,50)
```

```python
def myTrainTestSplit(X,y,test_size):
    idx = np.random.permutation(np.arange(X.shape[0]))
    tS = int(np.floor(test_size*X.shape[0]))
    testIdx = idx[:tS]
    trainIdx = idx[tS:]
    Xtrain = X[trainIdx]
    ytrain = y[trainIdx]
    Xtest = X[testIdx]
    ytest = y[testIdx]
    return Xtrain,Xtest,ytrain,ytest
```

```python
Xtrain,Xtest,ytrain,ytest = myTrainTestSplit(X,y,0.2)
```
