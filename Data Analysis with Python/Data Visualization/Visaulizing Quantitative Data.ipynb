**Visaulizing Quantitative Data**

- Quantitative data is all about numerical data.
- Data that consists of amounts or quantities
- Two types of charts that are commonly used to visualize quantitaive data is:
  - Histogsrams and Scatter plots.
  


For example in Airbnb listing in New York.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
import numpy as np
```

```python
listings = pd.read_csv("listings.csv")
```

```python
listings
```

> **Histogsrams**

Histograms are a type of graphical representation used to show the distribution of a continuous dataset. They provide insights into the frequency or count of data points falling within specified ranges or "bins" along a continuous axis

```python
plt.hist(listings["price"])
plt.xlabel("price(in US dollars)")
plt.show()
```

From the histograms above, the chart returns a price grouping into contiguous intervals called **bins**.

**NOTE: Meanig of word**
 - contiguous (adjective): next to or touching another, usually similar, thing.
 - interval (noun): a period between two events or times.
 
So  **contiguous intervals** or **bins** is the range of data between the start and end limits with the definition

So we will define a bins and will return a data is easiers to read and analyze from histograms

```python
plt.hist(listings["price"], bins = np.arange(0,1100,40))
plt.xlabel("price(in US dollars)")
plt.show()
```

> **Scatter plots**

A scatter plot is a type of data visualization that is used to display the relationship between two numerical variables. It is particularly useful for identifying patterns, trends, correlations, and outliers within the data. In a scatter plot, each data point is represented by a single dot, and the position of the dot is determined by the values of the two variables being plotted.

```python
plt.scatter(x= listings["price"], y= listings["number_of_reviews"])
plt.xlabel("price")
plt.ylabel("number of reviews")
plt.show()
```

So, we can restrict x axis fro can see a cotribution of chart.

( restrict (verb): to limit the movements or actions of someone, or to limit something and reduce its size or prevent it from increasing.)

```python
plt.scatter(x= listings["price"], y= listings["number_of_reviews"])
plt.xlabel("price")
plt.ylabel("number of reviews")
# Use for restrict the x or y axis
plt.xlim(0,1110)
plt.show()
```

The chart below is seem to clean but we can resize a point of scatter plots data by the s parameter(size) as show in below

```python
plt.scatter(x= listings["price"], y= listings["number_of_reviews"], s=4)
plt.xlabel("price")
plt.ylabel("number of reviews")
# Use for restrict the x or y axis
plt.xlim(0,1110)
plt.show()
```

At first glance at this plot, you will notice a trend. It appears that the listing with lower prices have more reviews, So visualizing your data can help you observe trends which can lead you to ask questions, which then informs a kind of further analysis that you may need to perform to answer those questions. And you can use things like histograms and scatter plots to visualize quantitative data.
