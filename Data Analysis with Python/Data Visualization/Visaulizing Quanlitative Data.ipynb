**Visaulizing Quanlitative Data**

- **Quanlitative** data is data that approximattes and characterizes. 
- It can be observed and recorded.
- It is non numerical in nature (not a number).
- Also know as categorical data. 
- It's subtypes are **nomminal data** and **ordinal data**.
- **Norminal data** is data that has no inherent order.
- **Ordinal data** is data that falls into ordered categories.

The data type in qualitative data visualization is non numerical in nature. This type of data can be usually collected through methods of observation. For example, a one to one interview conducting focus groups and similar methods.

> **Bar Chart**

A bar chart is a graphical representation used to display and compare the values of different categories or groups. It consists of bars of varying lengths or heights, with each bar representing a specific category or data point. Bar charts are particularly effective for showing the relationships between different data sets, identifying trends, and making comparisons. 

For example in Airbnb listing in New York.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
listings = pd.read_csv('listings.csv')
```

```python
listings
```

```python
# Need to focus "neighbourhood_group"

# From seaborn to create a bar chart
#               axis_name,           data from DataFrame
sn.countplot(x ="neighbourhood_group", data = listings)
# Use matplotlib to 
# show a car chart
plt.show()
```

```python
sn.barplot(x ="neighbourhood_group", y= "price", data = listings)
plt.show()
```

The bar above, The black line is basically are used to mark **confidence intervals**

- **NOTE**
  - **confidence intervals** are an iportant concept in data science nut we will talk about here later
