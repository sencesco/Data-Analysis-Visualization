**What is Data Cleaning?**

- Data cleaning is the process of combining through raw data and transforming it, so, that it's conductive to the analysis that follows.
- Data cleaning can detect and fix corrupt pt inaccurate records from a recoed set.
- Data cleaning can be time consuming, but it is a critical stage of the data science life cycle.

**What Data Cleaning can fix**

- Missing values
- Misspellings
- Duplicate rows
- inconsistence formats (Meaning many format in the data)
- Structure of overall data
- Extracting information from complex values 
- Unspecified unit conversions
- Interpretation of magnitude

**Summary**

In actually, The cleaning every aspect of a data set may take too long, but on the other hand, not cleaning a data set at all would lead to drawing inaccurate conclusions from the data.

So it is important to try to find a balance when you're cleaning data. Also, keep in mind that the decisions you make during data cleaning impacts all the data analysis you perform afterwards. If there's a column you choose not to clean in your data set, have caution when you're using that column in analysis.

Also know that it is helpful to explicitly keep track of what changes you make to your data set during the data cleaning so you can always refer or roll back to your notes when you are doing your analysis.
